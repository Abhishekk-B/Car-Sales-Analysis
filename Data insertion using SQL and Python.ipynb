```python
d_col=['Dealer_Name','Dealer_No ','Dealer_Region']
```

```python
dealer=df[d_col]
```

```python
dealer.shape
```

```python
dealers=dealer.drop_duplicates()
dealers.shape
```

```python
dealers['d_id']=dealers.index
dealers.shape
```

```python
df_temp = pd.merge(df, dealers, on=d_col)
df_temp.head()
```

```python
cars_col=['Company', 'Model', 'Engine', 'Transmission', 'Color','Price ($)','Body Style']
```

```python
car=df[cars_col]
car.shape
```

```python
cars=car.drop_duplicates()
cars.shape
```

```python
cars['c_id']=cars.index
cars.shape
```

```python
df_temp = pd.merge(df_temp, cars, on=cars_col)
df_temp.head()
```

```python
df_temp=df_temp.sort_values(by='Car_id').reset_index(drop=True)
df_temp.head()
```

```python
df_temp.info()
```

```python
from  mysql.connector import connect, Error
```

```python
try:
    with connect(
        host="localhost",
        user='root',
        password='mysql',
    ) as connection:
        query="CREATE DATABASE IF NOT EXISTS carsDB"
        cursor=connection.cursor()
        cursor.execute(query)
        print("Database is created.")       
except Error as e:
    print(e)
```

```python
cols=['Date','Annual Income','Price ($)','Phone','d_id','c_id']
for col in cols:
    df_temp[col]=df_temp[col].astype(str)
```

```python
try:
    with connect(
        host="localhost",
        user='root',
        password='mysql',
        database='carsDB'
    ) as connection:
        query="CREATE DATABASE IF NOT EXISTS carsDB"
        cursor=connection.cursor()
        cursor.execute("SELECT DATABASE();")
        db=cursor.fetchone()
        print("You are connected to the database",db)
        cursor.execute("DROP TABLE IF EXISTS carsDetails")
        print("Creating new table 'carsDetails'")
        query="""CREATE TABLE if not exists carsDetails(
        Car_id VARCHAR(50),
        Date DATETIME,
        CustomerName VARCHAR(50),
        Gender VARCHAR(10),
        AnnualIncome INT,
        Dealer_Name VARCHAR(50),
        Company VARCHAR(50),
        Model VARCHAR(50),
        Engine VARCHAR(50),
        Transmission VARCHAR(50),
        Color VARCHAR(50),
        Price INT,
        Dealer_No VARCHAR(50),
        BodyStyle VARCHAR(50),
        Phone VARCHAR(50),
        Dealer_Region VARCHAR(50),
        d_id int,
        c_id int)"""
        cursor.execute(query)
        print("Table created successfully!")        
        for i, row in df_temp.iterrows():
            query="INSERT INTO carsDetails VALUES (%s,%s,%s,%s,%s,%s,%s,%s,%s,%s,%s,%s,%s,%s,%s,%s,%s,%s)"
            cursor.execute(query,tuple(row))
            connection.commit()
        print('Record inserted successfully!')
except Error as e:
    print(e)
```
